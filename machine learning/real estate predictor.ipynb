# Train test splitting




```python
#using function
import numpy as np

def split_train_test(data,test_ratio):
    np.random.seed(42)
    shuffled= np.random.permutation(len(data))
    test_set_size = int(len(data)*test_ratio)
    test_indices=shuffled[:test_set_size]
    train_indices=shuffled[test_set_size:]
    return data.iloc[train_indices],data.iloc[test_indices]
```

# Scikit-learn design

Three types of objects
1. Estimators: Ex: Imputer which estimates a parameter based on dataset;has fit method which fits the datset and calculates internal parameters
2. Transformers:takes input and returns output based on learning from fit();has fit_transform()
3. Predictors:Ex:LinearRegression model has 2 functions fit and predict

# feature scaling

2 types feature scaling:
1. Min max scaling : from 0-1
value-min/(max-min)
scikitlearn provides class called MinMaxScaler
2. Standardization:
(value-mean)/std
scikitlearn provides class called Standard Scaler

# Creating pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
my_pipeline = Pipeline([
    ('imputer',SimpleImputer(strategy='median')),
                       ('std_scaler',StandardScaler()),
])
```

```python
housing_num_tr=my_pipeline.fit_transform(housing_tr)
```

```python
housing_num_tr
```

```python
housing_num_tr.shape
```

# Selecting a desired model for real estates

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble  import RandomForestRegressor
#model = LinearRegression()
#model = DecisionTreeRegressor()
model=RandomForestRegressor()
model.fit(housing_num_tr,housing_labels)
```

Evaluation of matrix

```python
from sklearn.metrics import mean_squared_error
housing_predictions = model.predict(housing_num_tr)
mse = mean_squared_error(housing_labels,housing_predictions)
rmse = np.sqrt(mse)
```

```python
rmse
```

# using better  evaluation technique - cross validation

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model,housing_num_tr,housing_labels,scoring = 'neg_mean_squared_error',cv=10)
#to max utility
rmse_scores=np.sqrt(-scores)
```

```python
rmse_scores
```

```python
def print_scores(scores):
    print("Scores are",scores)
    print("Mean:",scores.mean())
    print("Standard Deviation",scores.std())
```

```python
print_scores(rmse_scores)
```
